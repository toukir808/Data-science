```python
df.dropna(inplace=True)  # dropping the NA columns
```

```python
# converting column names  into list 
col=df.columns.to_list()
col
```

```python
# As per the research ther shoudl be no negative value should be present hence convering negtive value to positive
for i in col:
    df[i]=np.abs(df[i])
    # 
```

# Box plot before imputing outliers

```python
for i in col:
    df[i].plot(kind="box")
    plt.show()
```

```python
# counting the outliers and imputing them using IQR method 

def Handle_outlier(df):

    col=df.columns.to_list()

# removing Tendency and NSP and there outlier value are important
    for r in ["Tendency","NSP"]:
        col.remove(r)


    for i in col:
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 -Q1

        lower_bound =Q1 -1.5 *IQR
        upper_bound = Q3 +1.5 * IQR

        # Printing the outlier
        outlier_count= ((df[i] < lower_bound) | (df[i]>upper_bound)).sum()
        print(f"{i} : {outlier_count} outliers dfound ")

        # Imputing the outlier 
        df[i]=df[i].clip(lower_bound,upper_bound)
    return df

df= Handle_outlier(df)
```

```python
# Rounding below columns as there value should lie  [1,2,3]
cat=["Tendency","NSP"] 
for i in cat:
    df[i]=round(df[i],0)
    
```

```python
# convering the value of NSP to 3 
df.loc[df['NSP'] == 5,"NSP"] = 3
```
