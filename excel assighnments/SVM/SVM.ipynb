```python
from sklearn.preprocessing import LabelEncoder

df_encoded = data.copy()

le=LabelEncoder()
for col in cat_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])
```

```python
df_encoded
```

* Since all the faetures are non-ordinal we applied Label Encoding on each faeture.



```python
x=df_encoded.drop('class',axis=1)
y=df_encoded['class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42, k_neighbors=1)
X_resampled, y_resampled = sm.fit_resample(x_train, y_train)
```

```python
x_train
```

```python
y_resampled.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std_train  = scaler.fit_transform(X_resampled)
X_std_test  = scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_std_train,y_resampled)
```

```python
y_pred=svm.predict(X_std_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report,confusion_matrix
```

```python
print("Accuracy Score:",accuracy_score(y_test,y_pred))
print("Precision Score:",precision_score(y_test,y_pred))
print("Recall Score:",recall_score(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
print("Classification report",classification_report(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Greens')
```

```python
from sklearn.model_selection import  GridSearchCV
```

```python
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
grid = GridSearchCV(SVC(), param_grid, cv=3, scoring='accuracy',n_jobs=-1,verbose=3)
grid.fit(x_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

* The RBF kernel with C=1 and gamma='auto' gave the best accuracy (~83.8%) during model tuning.


```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_test_2D = pca.fit_transform(x_test)

plt.scatter(X_test_2D[:,0], X_test_2D[:,1], c=y_pred, alpha=0.7)
plt.title("SVM Classification Results (PCA 2D)")
plt.show()
```

```python
kernels = ['linear', 'poly', 'rbf']
for k in kernels:
    model = SVC(kernel=k)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    print(f"\nKernel: {k}")
    print("Accuracy:", accuracy_score(y_test, pred))
```

* Based on the results, both the linear and RBF kernels show similar performance with 66.75 % accuracy, while the polynomial kernel achieves the highest accuracy among the three with 69.25% accuracy.

Analyze SVM strengths and weaknesses for the Mushroom dataset based on EDA and visualization results.


* Strengths:

* Handles High-Dimensional Data Well: The mushroom dataset, after encoding, becomes high-dimensional. SVM can effectively find decision boundaries in such feature spaces.

* Good Performance with Clear Separation: The classes in the mushroom dataset (edible vs. poisonous) are well-separated, which aligns with SVM’s strength in finding maximum-margin hyperplanes.

* Flexibility via Kernels: Different kernels (linear, RBF, polynomial) allow SVM to adapt to both linear and non-linear patterns in the dataset.

* Robust to Overfitting in High-Dimensional Space: With proper regularization (C parameter), SVM maintains good generalization.



* Weaknesses:

* Computational Cost: SVM can be slow on large datasets. Even with 2,000 rows, non-linear kernels take more time compared to simpler models like Logistic Regression.

* Sensitive to Parameter Tuning: Kernel type, C, and gamma require careful tuning; poor choices can lead to underfitting or overfitting.

* Limited Probability Output: SVM does not naturally provide probability scores (needs extra computation with probability=True).

* Interpretability: The model’s decision boundary and feature influence are less interpretable compared to tree-based methods.

* Discuss practical implications of SVM in real-world classification tasks.

* Medical Diagnosis – SVM is effective in identifying diseases such as cancer (e.g., classifying tumor cells as malignant or benign) using patient medical records and imaging data. Its ability to work well with high-dimensional features makes it suitable for complex biomedical datasets.

* Text and Document Classification – SVM is frequently applied in spam detection, sentiment analysis, and topic categorization due to its efficiency in handling sparse and high-dimensional data from text processing.

* Image Recognition – It is used for tasks like face detection, handwriting recognition, and object classification, where feature dimensions are large, and clear decision boundaries are required.

* Bioinformatics – SVM plays a crucial role in gene classification, protein structure prediction, and classification of bio-sequences, where datasets are often small but with high feature counts.

* Financial Risk Assessment – Banks and financial institutions use SVM to classify credit risk levels, detect fraud, and predict loan defaults based on customer profiles.
